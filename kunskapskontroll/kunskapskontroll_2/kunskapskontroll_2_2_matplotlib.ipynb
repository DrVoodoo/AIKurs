### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

```python
# One way is to rely on pyplot to implicitly create figures and axes and then use pyplot functions for plotting
# The other recommended way is to use an OO oriented approach and explicitly create figures and axes and call methods on them
```

### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

```python
# Figure contains everything as axes, artists, nested subfigures and more. It is the root object
# Axes is an artist attached to a figure where it has an area it can plot data to.
# An axes contain axis objects. Normally an axes object has 2 or 3 axis object in it for plotting in 2 (xy) or 3 dimensions (xyz)
# An axis is an artist handling the ticks, scale and limits along one direction in the figure. Ie x-axis or y axis.
# Basically everything visible in a figure is an artist, texts, plots, axis etc
```

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
data = np.linspace(-4,4,50)
fig, ax = plt.subplots(figsize=(3, 3), layout='constrained')
ax.plot(data, data**2, label='quadratic')
ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.set_title("y = x^2")
```

### Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python
# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())
```

```python
fig, axd = plt.subplot_mosaic([['scatterplot', 'barplot']], figsize=(6,3), layout='constrained')

axd['scatterplot'].set_title('Scatter plot')
axd['scatterplot'].scatter(x, y)

axd['barplot'].set_title('Bar plot')
axd['barplot'].bar(names, values)
```
